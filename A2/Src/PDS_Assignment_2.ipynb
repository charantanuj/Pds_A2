```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

Data=pd.read_csv("/content/train.csv")
```

a) Look for the missing values in all the columns and either impute them (replace with mean,
median, or mode) or drop them.

```python
Data.isnull().sum()
```

```python
Data.fillna(Data.mean(), inplace=True)
Data.drop('New_Price', axis=1, inplace=True)
```

```python
Data.isnull().sum()
```

```python
num_cols = Data.select_dtypes(include=['float64', 'int64']).columns
Data[num_cols] = Data[num_cols].fillna(Data[num_cols].mean())

cat_cols = Data.select_dtypes(include=['object']).columns
Data[cat_cols] = Data[cat_cols].fillna(Data[cat_cols].mode().iloc[0])

Data.to_csv('/content/cleanData.csv')

Data.isnull().sum()
```

b) Remove the units from some of the attributes and only keep the numerical values (for
example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from
“New_price”).

```python
Data['Mileage'] = Data['Mileage'].str.extract('(\d+\.\d+)').astype(float)
Data['Engine'] = Data['Engine'].str.replace(' CC', '').astype(float)
Data['Power'] = Data['Power'].str.replace(' bhp', '').astype(float)

Data.fillna(Data.mean(), inplace=True)

Data.to_csv('/content/Units_Removed_Data.csv')

Data
```

 C) Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot encoded value.

```python
# Mapping Fuel_Type
FT_map = {'Diesel': 0, 'Petrol': 1, 'Electric': 2}
Data['Fuel_Type'] = Data['Fuel_Type'].map(FT_map)

# Mapping Transmission
T_map = {'Manual': 0, 'Automatic': 1}
Data['Transmission'] = Data['Transmission'].map(T_map)

# Mapping Owner_Type
OT_map = {'First': 0, 'Second': 1, 'Third': 2, 'Fourth & Above': 3}
Data['Owner_Type'] = Data['Owner_Type'].map(OT_map)

Data.to_csv('/content/Encoded_Data.csv')

Data
```

D) Create one more feature and add this column to the dataset (you can use mutate function in
R for this). For example, you can calculate the current age of the car by subtracting “Year” value
from the current year.

```python
Cur_Year = pd.Timestamp.now().year
Data['Current_Age'] = Cur_Year - Data['Year']

Data.to_csv('/content/Feature_Added_Data.csv')

Data
```

e) Perform select, filter, rename, mutate, arrange and summarize with group by operations (or
their equivalent operations in python) on this dataset.

```python
# Select specific columns
Selected_data = Data[['Location','Year', 'Price', 'Mileage', 'Engine', 'Seats']]

# Filter
Filtered_data = Data[Data['Year'] > 2017]

# Rename columns
Renamed_data = Data.rename(columns={'Engine': 'Engine_CC', 'Seats': 'Number_of_Seats','Price': 'Price_Lakhs','Power': 'Power_bhp'})

# Mutate
Data['Engine'] = Data['Engine'] / 1000  # Converting CC to Liters

# Arrange data
arranged_data = Data.sort_values(by='Price', ascending=False)

# Summarize with group by operations
summary_data = Data.groupby('Location')['Price'].mean()

Data.to_csv('/content/Operations_Performed_Data.csv')

Data
```
